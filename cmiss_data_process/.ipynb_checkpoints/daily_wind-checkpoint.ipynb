```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif']=['SimHei'] #用来正常显示中文标签
plt.rcParams['axes.unicode_minus']=False #
import netCDF4 as nc
import glob
```

```python
#把2400个站点的经纬度读入
file0= pd.read_csv('/home/liuli/OBS_ZZ/daily/china_stations2540.txt',sep='\t',header=None)
df0=pd.DataFrame(file0)
print(df0)
print(df0.columns)
```

```python
data_path='/home/liuli/OBS_ZZ/daily/wind/day/'
mulu = glob.glob(data_path+'*')
mulu.sort()
df_all=pd.DataFrame()
```

```python
for file in mulu:
    
    file0= nc.Dataset(file)
    wind=file0.variables['win']
    wind=np.array(wind)
    wind=np.where(wind<1000,wind,np.nan)
    wind=pd.DataFrame(wind)
    df_all=pd.concat([df_all,wind],ignore_index=True)
    del wind
    del file0
df_all.columns=df0[0].values

print(df_all)
```

```python
name = 'selected_station.txt'
name2 = 'valid_ratio.txt'
#count=0
f=open(name,'w')
f2=open(name2,'w')
for co in df_all.columns: 
    num=df_all[co].count()
    rate=num/len(df_all)
    if rate>=0.5: 
        #count+=1
        line=[]
        line2=[]
        line=str(co)+'\t'+str(df0[df0[0].isin([co])][1].values[0])+'\t'+str(df0[df0[0].isin([co])][2].values[0])+'\n'
        line2=str(co)+'\t'+str(rate)+'\n'
        f.write(line)
        f2.write(line2)
        #print(count,co,rate)
f.close()
f2.close()
```
