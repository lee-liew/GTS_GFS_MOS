```python
#尝试统计每个站点气象要素的缺测情况（以降水缺测统计为主）
cols=['STATION', 'YEAR', 'MONTH','DAY','HOUR','LATITUDE','LONGITUDE','ELEVATION','N','WD','WS','AT','TD','RH','LP','DP03','NH','H1','RAIN01','RAIN02','RAIN03','RAIN06','RAIN09','RAIN12','RAIN24','MIN_AT','ALR','RAIN032','RAIN062','RAIN122','RAIN242']

file0= pd.read_csv('/home/liuli/GTS_GFS_MOS/surf2018.csv',usecols=cols,sep=',',encoding='ISO-8859-1')
df0=pd.DataFrame(file0)
print(df0)

#df0.count()
#df0['STATION'].value_counts()

#####1.去掉重复的行
df1=df0.drop_duplicates(subset = ['STATION','YEAR','MONTH','DAY','HOUR'],keep='last',inplace=False)
#####2.合并年月日为时间戳，接下来要照年月日小时重新排列数据
periods = pd.PeriodIndex(year=df1['YEAR'], month=df1['MONTH'], day=df1['DAY'], hour=df1['HOUR'],freq="H")
#print(periods)
df1.insert(0,'timeStamp', periods)
#
print(df1)
```

```python
#申明一个不重复站点的df
df_qc=df1.drop_duplicates(subset = ['STATION'],keep='last',inplace=False)
print(df_qc)
```

```python
####3.统计每个站点的缺测情况，根据数据完好状况来挑选站点。
#print(num_sta)
#此事件用于循环里面的日期补齐
start='2019-01-01 00:00:00'
end_of_db = '2020-01-01 00:00:00'
date_index = pd.date_range(start, end_of_db, freq = '3h')
date_df = pd.DataFrame(date_index,columns=['timeStamp'])
#print(date_df)
#print(date_df.dtypes)
df1['timeStamp'] = df1['timeStamp'].dt.to_timestamp() #必须要转换timestamp，把object转换为datetime64[ns]类型，为merge做准备
#df1['timeStamp']=pd.to_datetime(df1['timeStamp']) #用了这一句报错，不知道为什么
#print(df1.dtypes)
#cols_data是用于统计有效数据的列
cols_data=['N','WD','WS','AT','TD','RH','LP','DP03','NH','H1','RAIN01','RAIN02','RAIN03','RAIN06','RAIN09','RAIN12','RAIN24','MIN_AT','ALR','RAIN032','RAIN062','RAIN122','RAIN242']

name = 'statistics_valid_data_2.txt'
f=open(name,'w')

for i in df_qc['STATION']:   
    df2=df1[df1['STATION'].isin([i])] #先挑出这个站号的行    
    #df2=df2.sort_values(by='timeStamp')#把这个站点按照时间顺序排列
    df2=df2.reset_index(drop=True)
    #print(df2)
    # merge日期补充完整数据
    df2 = pd.merge(date_df,df2, on=['timeStamp'], how='left')    
    #print(df2)
    for co in cols_data:        
        num=df2[co].count()
        rate=num/len(df2)
        if rate>=0.5: 
            line=[]
            line=str(i)+'\t'+str(co)+'\t'+str(rate)+'\n'
            f.write(line)
        #把每种要素的有效率结果存入文件
        #print(i,co,rate)
    del df2
```
